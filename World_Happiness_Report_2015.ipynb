```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df =pd.read_csv('/content/2015.csv')
```

```python
df.head()
```

# 1. Distribution of Happiness Scores for Each Country

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.histplot(df['Happiness Score'], bins=20, kde=True)
plt.title('Distribution of Happiness Scores')
plt.xlabel('Happiness Score')
plt.ylabel('Frequency')
plt.show()
```

# 2. Country with the Highest Happiness Score

```python
highest_happiness_country = df.loc[df['Happiness Score'].idxmax()]
print(f"Country with highest happiness score: {highest_happiness_country['Country']}")
print(f"Happiness score: {highest_happiness_country['Happiness Score']}")
```

# 3. Correlation between GDP per capita and Happiness Score

```python
correlation = df['Economy (GDP per Capita)'].corr(df['Happiness Score'])
print(f"Correlation between GDP per capita and happiness score: {correlation}")
```

# 4. Distribution of Factors Contributing to Happiness

```python
factors = ['Economy (GDP per Capita)', 'Family', 'Health (Life Expectancy)', 'Freedom', 'Trust (Government Corruption)', 'Generosity']
plt.figure(figsize=(12, 8))
sns.boxplot(data=df[factors])
plt.title('Distribution of Factors Contributing to Happiness')
plt.xticks(rotation=45, ha='right')
plt.show()
```

# 5. Region with the Highest Average Happiness Score

```python
region_happiness = df.groupby('Region')['Happiness Score'].mean().sort_values(ascending=False)
print("Region with highest average happiness score:")
print(region_happiness.head(1))
```
