```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/Salaries.csv')
```

```python
df
```

Find The Occurrence Of the Employee Names

```python
employee_name_counts = df['EmployeeName'].value_counts().head(5)
print(employee_name_counts)
```

 Find The Number Of Unique Jobs Titles

```python
unique_job_titles = df['JobTitle'].nunique()
print(unique_job_titles)
```

 Find The Total Number Of Job Titles Contain captain

```python
captain_jobs = df[df['JobTitle'].str.contains('Captain', case=False)]['JobTitle'].count()
print(captain_jobs)
```

Display All the Employee Names from Fire Department

```python
fire_department_employees = df[df['JobTitle'].str.contains('Fire Department', case=False)]['EmployeeName']
print(fire_department_employees)
```

 Find The Minimum, Maximum and Average BasePay

```python
min_basepay = df['BasePay'].min()
max_basepay = df['BasePay'].max()
avg_basepay = df['BasePay'].mean()

print("Minimum BasePay:", min_basepay)
print("Maximum BasePay:", max_basepay)
print("Average BasePay:", avg_basepay)
```

 Replace ‘Not provided’ in EmployeeName Column to Nan

```python
df['EmployeeName'] = df['EmployeeName'].replace('Not provided', np.nan)
```

```python
df
```

 Drop The Rows Having 5 Missing Values

```python
df = df.dropna(thresh=len(df.columns) - 4)
# Keep rows with at least (total columns - 4) non-missing values
```

```python
df
```

Find The Job Title Of ALBERT PARDINI

```python
albert_job_title = df[df['EmployeeName'] == 'ALBERT PARDINI']['JobTitle'].iloc[0]
print(albert_job_title)
```

 How Much ALBERT PARDINI makes (include Benefits)

```python
albert_total_pay = df[df['EmployeeName'] == 'ALBERT PARDINI']['TotalPayBenefits'].iloc[0]
print(albert_total_pay)
```

Display Name Of The Person Having The Highest Basepay

```python
highest_basepay_person = df.loc[df['BasePay'].idxmax(), 'EmployeeName']
print(highest_basepay_person)
```

 Find The average Basepay of all Employees per Year

```python
avg_basepay_per_year = df.groupby('Year')['BasePay'].mean()
print(avg_basepay_per_year)
```

Find Average Basepay Of All Employee Per Job Title

```python
avg_basepay_per_job = df.groupby('JobTitle')['BasePay'].mean()
print(avg_basepay_per_job)
```

Find The Average Basepay of All Employee Having job Title Accountant

```python
accountant_avg_basepay = df[df['JobTitle'] == 'Accountant']['BasePay'].mean()
print(accountant_avg_basepay)
```

Five Most Common Jobs

```python
most_common_jobs = df['JobTitle'].value_counts().head(5)
print(most_common_jobs)
```
