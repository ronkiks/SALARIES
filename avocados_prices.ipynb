```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/avocado.csv')
```

```python
df.head()
```

# 1.  What is the average price of avocados?

```python
average_price = df['AveragePrice'].mean()
print(f"The average price of avocados is: {average_price}")
```

# 2. What is the distribution of the average price by region?

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='region', y='AveragePrice', data=df)
plt.xticks(rotation=90)
plt.title('Distribution of Average Avocado Price by Region')
plt.show()
```

# 3. Which region has the highest and lowest average price?

```python
region_prices = df.groupby('region')['AveragePrice'].mean().sort_values()
highest_region = region_prices.index[-1]
lowest_region = region_prices.index[0]
print(f"Region with highest average price: {highest_region}")
print(f"Region with lowest average price: {lowest_region}")
```

# 4. What is the correlation between the total volume and the average price?

```python
correlation = df['Total Volume'].corr(df['AveragePrice'])
print(f"Correlation between total volume and average price: {correlation}")
```

# 5. What is the distribution of the total volume by year?

```python
plt.figure(figsize=(10, 5))
sns.histplot(x='year', y='Total Volume', data=df)
plt.title('Distribution of Total Volume by Year')
plt.show()
```
